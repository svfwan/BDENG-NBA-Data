```python
from kafka import KafkaProducer
import csv
import json
```

```python
bootstrap_servers = 'kafka:9092'
topic_name = 'nba-WEB-data'

producer = KafkaProducer(
    bootstrap_servers=bootstrap_servers,
    key_serializer=lambda x: str(x).encode('utf-8'),
    value_serializer=lambda x: json.dumps(x).encode('utf-8')
)

# Read data from CSV files
seasons = ['1996', '1997', '2005', '2006', '2007', '2011', '2012', '2013']
for season in seasons:
    filename = f"{season}.csv"
    with open(filename, 'r') as file:
        csv_reader = csv.DictReader(file)
        for row in csv_reader:
            # Create a message with the year as the key and the row data as the value
            player_data = {
                'season': season,
                'data': row
            }
            producer.send(topic_name, value=player_data)
            print(f"Sent message for Season: {season}: {row}")

producer.flush()
```
